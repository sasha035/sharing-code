```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

df1 = pd.read_csv('survey1.csv')
df2 = pd.read_csv('steps1.csv', sep=';')
df1.head()
```

```python
df2.head()
```

```python
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head()
```

```python
df.info() #Get information on the variables
```

```python
df.fillna(0)
```

```python
df.mean(axis=1)
```

```python
df["Mean_steps"] = df.mean(axis=1)
df.head()
```

```python
#plotting the means of all participants
sns.distplot(df.Mean_steps)
```

```python
#Graph 1 distribution mean steps per day
sns.kdeplot(df.Mean_steps)
```

```python
#Graph 2 distribution mean steps per day
sns.boxplot(df["Mean_steps"])
plt.title('Mean of total steps all participants')
plt.xlabel('Mean steps')
plt.show()
```

```python
#Create a boxplot with 2 variables: 1 qualitative (gender), 1 quantitative (mean steps per day)
sns.catplot(x='Mean_steps', y='gender', kind='box', data=df)
plt.title('Mean of total steps all participants')
plt.xlabel('Mean steps')
plt.show()
```

```python
import seaborn as sns #this is the plotting library I'll be using 
import pandas as pd #"as pd" means that we can use the abbreviation in commands
import matplotlib.pyplot as plt #we need Matplotlib for setting the labels in the Seaborn graphs

df1 = pd.read_csv('survey1.csv')
df2 = pd.read_csv('steps1.csv', sep=';')
df1.head()
```

```python
df2.head()
```

```python
# add the average steps per day per id, I do this for the plot later.
df2['Average_steps'] = df2.mean(numeric_only=True, axis=1)
df2.head()
```

```python
df = pd.merge(df1, df2, on = 'id') #'id' is the common identifier
df.head()
```

```python
df.info() #Get information on the variables
```

```python
df.fillna(0)
```

```python
#remove data with 65000 steps or more (+/- 50km)
def above_65000(x):
    if(x > 65000.0): 
        return float('NaN')
    else: 
        return x
df['Average_steps'] = df['Average_steps'].apply(above_65000)

df.head()
```

```python
#Graph 1 distribution mean steps per day
sns.kdeplot(df.Average_steps)
```

```python
df['Average_steps'].name = 'Average steps' #Labeling the variable for plotting
Average_steps = df['Average_steps'].dropna() #Selecting the size column. I need to drop the NA's, otherwise Seaborn won't plot
sns.distplot(Average_steps, kde=False) 
plt.title('Distribution of the average steps')
plt.show()
```

```python
#Graph 2
sns.boxplot(df.Average_steps)
plt.title('Average of steps per day')
plt.xlabel('Steps')
plt.show()
```

```python
#Make summary of statistics (measure of central tendency and variation, e.g., mean, median, variance)
mean_size = df['Average_steps'].mean()
median_size = df['Average_steps'].median()
variance = df['Average_steps'].var()

print(f'The median daily steps is: {median_size} steps')
print(f'The mean daily steps is: {mean_size} steps')
print(f'The variance daily steps is: {variance} steps')
```

## Discription
The mean of the distribution of the data on daily steps is higher than the median. Meaning, right skewed. The distribution seems normal, having only a few outliers above 15000 steps per day. These outliers may be interesting edge cases. Therefore, I did not remove them. Also, because this is a model of the whole group of participants and not a selecter group of professionals.
